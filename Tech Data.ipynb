```python
import pandas as pd
```

# Data Cleaning

```python
# Load files
tech_data=pd.read_csv("Resources/Tech_data.csv")
tech_data.head()
```

```python
# Check data types
tech_data.dtypes
```

```python
# Check Nulls
tech_data.isnull().sum()
```

```python
X=tech_data.drop(columns='date')
X
```

# Preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler().fit(X)
```

```python
X_scaled=scaler.transform(X)
X_scaled
```

```python
X_scaled_df=pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.head()
```

```python
X_scaled_df.describe()
```

```python
X_scaled_df.to_csv("Scaled_tech_data.csv")
```
